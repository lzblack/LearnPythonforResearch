### Generate a new dataframe with the number of complaints per "Borough"

```python
df_complaints_per_borough = df_311.groupby('Borough').size().to_frame('Number of Complaints').reset_index()
df_complaints_per_borough
```

### Add a new column to the dataframe you have created above (which contains 1 row per Borough) with the population for each borough

Hint: use `df.column.str.title()`

```python
df_complaints_per_borough['Borough']=df_complaints_per_borough['Borough'].str.title()

df_complaints_per_borough
```

```python
# download the data from https://data.cityofnewyork.us/City-Government/NYC-Population-by-Borough/h2bk-zmw6
population_filename = 'NYC_Population_by_Borough.csv'

df_population = pd.read_csv(join(data_path, population_filename), low_memory=False)
df_population
```

```python
df_complaints_per_borough = pd.merge(df_complaints_per_borough,
                                    df_population,
                                    how='left',
                                    on='Borough')
df_complaints_per_borough
```

### Using the original dataframe, which borough has the most "Noise - ....." complaints?

Hint: first generate a column that is 1 if a complaint is "Noise - ...." 

```python
df_311['Complaint Type - Noise'] = np.where(df_311['Complaint Type'].str.startswith('Noise -' ), 
                                                         1, 0)
df_311[df_311['Complaint Type - Noise']==1].groupby('Borough').size()\
    .sort_values(ascending=False).to_frame('Number of Noise Complaints')
```

### For the complaints that are closed, what is the average number of hours wait time between open and close for each agency?

**Note:** This one is more complicated

Hint: you can to convert a timedelta column to the number of minutes by doing this:  
> df_closed['wait'].apply(lambda x: x / np.timedelta64(1, 'h'))

```python
df_closed = df_311[df_311['Status'] == 'Closed']
df_closed['wait'] = (pd.to_datetime(df_closed['Closed Date']) - 
                     pd.to_datetime(df_closed['Created Date']))/ np.timedelta64(1, 'h')
df_closed.sample(5)
```

```python
df_closed.groupby('Agency')['wait'].mean().sort_values(ascending=False)
```
